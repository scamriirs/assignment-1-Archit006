```python
# Q1: List Manipulation - Shopping Cart

# Step 1: Create an empty list
shopping_cart = []

# Step 2: Ask the user to input three items one by one
for i in range(3):
    item = input(f"Enter item {i+1}: ")
    shopping_cart.append(item)

# Step 3: Print the final list
print("Final Shopping Cart:", shopping_cart)

# Step 4: Remove the second item (index 1)
if len(shopping_cart) >= 2:
    removed_item = shopping_cart.pop(1)
    print(f"Removed item: {removed_item}")

# Step 5: Print the updated list
print("Updated Shopping Cart:", shopping_cart)
```

```python
# Q3: Dictionary - Phone Book

# Step 1: Create an empty dictionary
phone_book = {}

# Step 2: Add three key-value pairs (name: phone number)
phone_book["Alice"] = "9876543210"
phone_book["Bob"] = "9123456780"
phone_book["Charlie"] = "9988776655"

# Step 3: Print the entire dictionary
print("Phone Book:", phone_book)

# Step 4: Ask the user for a name
name = input("Enter a name to search: ")

# Step 5: Print the phone number or handle if not found
if name in phone_book:
    print(f"{name}'s phone number is: {phone_book[name]}")
else:
    print(f"Sorry, {name} not found in phone book.")
```

```python
# Q7: Membership Testing (Lists & Dictionaries)

# Step 1: Create a list of fruits
fruits = ['apple', 'banana', 'mango']

# Step 2: Ask the user for a fruit
fruit_name = input("Enter a fruit name: ").lower()

# Step 3: Check if the fruit is in the list
if fruit_name in fruits:
    print(f"Yes, we have {fruit_name}s!")
else:
    print(f"Sorry, {fruit_name} is not in the list.")

# Step 4: Dictionary with fruit prices
prices = {'apple': 1.0, 'banana': 0.5, 'orange': 0.75}

# Step 5: Check if the fruit is in dictionary keys
if fruit_name in prices:
    print(f"{fruit_name.capitalize()} is available in the price list at ${prices[fruit_name]}")
else:
    print(f"{fruit_name.capitalize()} is not listed in the price dictionary.")
```

```python
# Q8: Combining Lists and Dictionaries - Gradebook

# Step 1: Create the dictionary with student names and their scores
grades = {
    "Alice": [85, 90, 78],
    "Bob": [92, 88, 95],
    "Charlie": [70, 75, 80]
}

# Step 2: Choose a student (e.g., Alice)
student_name = "Alice"

# Step 3: Calculate the average
if student_name in grades:
    scores = grades[student_name]
    average = sum(scores) / len(scores)
    print(f"{student_name}'s average score: {average:.2f}")
else:
    print(f"Student {student_name} not found in gradebook.")
```

```python
# Q9: Tuple Unpacking - Coordinates

# Step 1: Create a tuple representing a 2D coordinate
point = (5, -3)

# Step 2: Unpack the tuple into x and y
x, y = point

# Step 3: Print the values of x and y
print("x-coordinate:", x)
print("y-coordinate:", y)

# Step 4: Calculate distance from origin
distance = (x**2 + y**2) ** 0.5
print("Distance from origin:", distance)
```

```python
# Q10: Comprehensive Practice - Restaurant Order

# Step 1: Create a dictionary for the menu
menu = {
    "pizza": 250,
    "burger": 150,
    "pasta": 200,
    "salad": 120,
    "coffee": 80
}

# Step 2: Create an empty order list
order = []

# Step 3: Ask the customer for two items
print("Menu:", menu)
for i in range(2):
    item = input(f"Enter item {i+1} from the menu: ").lower()
    if item in menu:
        order.append(item)
    else:
        print(f"Sorry, {item} is not available.")

# Step 4: Calculate the total cost
total = sum(menu[item] for item in order)

# Step 5: Print the total bill
print("Your order:", order)
print("Total Bill Amount:", total)
```
